```python
def nullscan(df_check):
    df_nulls = df_check.isna()
    nulls_per_col = df_nulls.sum(axis=0)
    nulls_per_col /= len(df_check.index)
    with plt.style.context('ggplot'):
        fig, (ax2) = plt.subplots(nrows=1, ncols=1, sharex=True, figsize=(6, 8))
        nulls_per_col.plot(kind='bar', color='c', x=nulls_per_col.values, 
                           y=nulls_per_col.index, ax=ax2, width=1, linewidth=1, 
                           edgecolor='black', align='edge', label='Null value rate')
        ax2.set_ylim((0,1))
        labels=df_check.columns
        ticks = np.arange(0.5, len(labels))
        ax2.xaxis.set(ticks=ticks, ticklabels=labels)
        #ax2.spines['top'].set_color('black')
        #ax2.spines['right'].set_color('black')

        na_ticks = ticks[(nulls_per_col > 0) & (nulls_per_col < 0.05)]
        if (len(na_ticks) > 0):
            ax2.plot(na_ticks, [0,]*len(na_ticks), 's', c='c', markersize=10, 
                     label='Very few missing values')
    
        fig.suptitle('Null Value Rate per Column', fontsize=20, y=1.05)
        ax2.legend()
        fig.tight_layout()
        plt.show()
nullscan(df)
```

```python
new_df = df
```

```python
df_knn = df.loc[:,['price','points','country','taster_name']]
dum_var = pd.get_dummies(df_knn[['country','taster_name']], prefix=['country','taster_name'])
df_knn = pd.concat([df_knn.drop(['country','taster_name'], axis=1), dum_var], axis =1)
```

```python
#Only chossing those values which doesn't have any null values
df_knn_non = df_knn.loc[df_knn['price'].notnull(), :]

x_non = df_knn_non.drop('price', axis =1)
y_non = df_knn_non['price']

df_knn_null = df_knn.loc[df_knn['price'].isnull(), :]
x_null = df_knn_null.drop('price', axis=1)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV
knn = KNeighborsRegressor()

n = {'n_neighbors' : [1,5,9]}
gs = GridSearchCV(knn, n)
gs.fit(x_non, y_non)
knn = gs.best_estimator_
```

```python
df_knn_non_null = new_df.loc[new_df['price'].notnull(),:]
df_knn_null_value = new_df.loc[new_df['price'].isnull(),:]

df_knn_null_value = df_knn_null_value.drop('price', axis =1)
df_knn_null_value['price'] = knn.predict(x_null)

new_df = pd.concat([df_knn_non_null,df_knn_null_value], axis = 0, ignore_index= True)

new_df = new_df.sample(frac=1).reset_index(drop = True)

nullscan(new_df)
```
