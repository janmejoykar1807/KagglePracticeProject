```python
def nullscan(df_check):
    df_nulls = df_check.isna()
    nulls_per_col = df_nulls.sum(axis=0)
    nulls_per_col /= len(df_check.index)
    with plt.style.context('ggplot'):
        fig, (ax2) = plt.subplots(nrows=1, ncols=1, sharex=True, figsize=(6, 8))
        nulls_per_col.plot(kind='bar', color='c', x=nulls_per_col.values, 
                           y=nulls_per_col.index, ax=ax2, width=1, linewidth=1, 
                           edgecolor='black', align='edge', label='Null value rate')
        ax2.set_ylim((0,1))
        labels=df_check.columns
        ticks = np.arange(0.5, len(labels))
        ax2.xaxis.set(ticks=ticks, ticklabels=labels)
        #ax2.spines['top'].set_color('black')
        #ax2.spines['right'].set_color('black')

        na_ticks = ticks[(nulls_per_col > 0) & (nulls_per_col < 0.05)]
        if (len(na_ticks) > 0):
            ax2.plot(na_ticks, [0,]*len(na_ticks), 's', c='c', markersize=10, 
                     label='Very few missing values')
    
        fig.suptitle('Null Value Rate per Column', fontsize=20, y=1.05)
        ax2.legend()
        fig.tight_layout()
        plt.show()
nullscan(df)
```

### KNN imputation Method

```python
new_df = df
```

```python
df_knn = df.loc[:,['price','points','country','taster_name']]
dum_var = pd.get_dummies(df_knn[['country','taster_name']], prefix=['country','taster_name'])
df_knn = pd.concat([df_knn.drop(['country','taster_name'], axis=1), dum_var], axis =1)
```

```python
#Only chossing those values which doesn't have any null values
df_knn_non = df_knn.loc[df_knn['price'].notnull(), :]

x_non = df_knn_non.drop('price', axis =1)
y_non = df_knn_non['price']

df_knn_null = df_knn.loc[df_knn['price'].isnull(), :]
x_null = df_knn_null.drop('price', axis=1)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV
knn = KNeighborsRegressor()

n = {'n_neighbors' : [1,5,9]}
gs = GridSearchCV(knn, n)
gs.fit(x_non, y_non)
knn = gs.best_estimator_
```

```python
df_knn_non_null = new_df.loc[new_df['price'].notnull(),:]
df_knn_null_value = new_df.loc[new_df['price'].isnull(),:]

df_knn_null_value = df_knn_null_value.drop('price', axis =1)
df_knn_null_value['price'] = knn.predict(x_null)

new_df = pd.concat([df_knn_non_null,df_knn_null_value], axis = 0, ignore_index= True)

new_df = new_df.sample(frac=1).reset_index(drop = True)

nullscan(new_df)
```

### We are spliting our points into 5 segments of ratings.

### We are spliting the points into 5 different points or categories to visualise the points better.

```python
def transform_points(points):
    if points < 84:
        return 1
    elif points >= 84 and points < 88:
        return 2 
    elif points >= 88 and points < 92:
        return 3 
    elif points >= 92 and points < 96:
        return 4 
    else:
        return 5

#Applying transform method and assigning result to new column "ratings_on_points"
new_df = new_df.assign(ratings_on_points = new_df['points'].apply(transform_points))
new_df.head()
```

```python
dp = new_df[['country','ratings_on_points']]
dp.head()
```

### In the below plot:- 

* we are trying to find the top 10 countries, which has maximum count on on ratings. Or 
* in-directly saying the country which sells most wines based on no. of ratings.

```python
grouped = dp.groupby(['country', 'ratings_on_points']).size().unstack(fill_value=0)
top10 = grouped.sum(axis=1).sort_values(ascending=False)[:10]
grouped.loc[top10.index].plot(kind='bar', stacked=True)

plt.xlabel('Country Grouped by Rating')
plt.ylabel('Total Count')
plt.title('Stacked Barplot of Countries by Rating')

plt.show()
```

### In the below plot:-
* We are trying see which countries are selling wine for which of the following category of rating the most.

```python
grouped_data = dp.groupby(['country', 'ratings_on_points']).size().reset_index(name='counts')

# Pivot the data to create a matrix of counts for each country and rating
pivoted_data = grouped_data.pivot(index='ratings_on_points', columns='country', values='counts')

# Sort the countries by their total count across all ratings
sorted_countries = pivoted_data.sum(axis=0).sort_values(ascending=False).head(10).index.tolist()

# Filter the data to only include the top 10 countries
filtered_data = pivoted_data[sorted_countries]

# Plot a stacked bar chart
filtered_data.plot(kind='bar', stacked=True)

# Set the title and axis labels
plt.title('Stacked Bar Chart of Ratings by Country')
plt.xlabel('Rating')
plt.ylabel('Total Count')

# Show the plot
plt.show()
```
